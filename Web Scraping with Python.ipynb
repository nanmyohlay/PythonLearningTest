```python
res = requests.get('http://quotes.toscrape.com')
```

```python
#res.text
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
```

```python
soup
```

```python
soup.select('.author')
```

```python
authors = set()
for name in soup.select('.author'):
    authors.add(name.text)
```

```python
authors
```

```python
quotes =[]
for quote in soup.select('.text'):
    quotes.append(quote.text)
```
